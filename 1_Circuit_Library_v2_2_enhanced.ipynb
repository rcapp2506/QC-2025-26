## ðŸ“š Overview

Qiskit's circuit library provides pre-built quantum circuits.

### Categories:
- Standard gates
- Boolean logic
- Arithmetic
- State preparation

```python
from qiskit import QuantumCircuit
from qiskit.circuit.library import (
    QFT, GroverOperator,
    ZZFeatureMap, RealAmplitudes,
    EfficientSU2, TwoLocal,
    MCXGate, MCXVChain
)
import numpy as np
import matplotlib.pyplot as plt

print('âœ“ Imports successful')
```

## Quantum Fourier Transform

```python
# QFT for 3 qubits
qft = QFT(3)
qft.draw('mpl')
```

```python
print('QFT circuit:')
qft.decompose().draw('mpl')
```

## Grover Operator

```python
# Create oracle (mark state |11âŸ©)
oracle = QuantumCircuit(2)
oracle.cz(0, 1)

# Create Grover operator
grover_op = GroverOperator(oracle)
print('Grover operator:')
grover_op.decompose().draw('mpl')
```

## Feature Maps for QML

```python
# ZZ Feature Map
feature_map = ZZFeatureMap(feature_dimension=3, reps=2)
print('ZZ Feature Map:')
feature_map.decompose().draw('mpl', fold=-1)
```

## Variational Forms

```python
# Real Amplitudes ansatz
ansatz = RealAmplitudes(num_qubits=3, reps=2)
print('Real Amplitudes ansatz:')
ansatz.decompose().draw('mpl', fold=-1)
```

## Multi-Controlled Gates

```python
# Multi-controlled X gate
mcx = MCXGate(num_ctrl_qubits=3)
qc_mcx = QuantumCircuit(4)
qc_mcx.append(mcx, range(4))
print('Multi-controlled X (Toffoli with 3 controls):')
qc_mcx.decompose().draw('mpl')
```

## Custom Library Circuits

```python
# Create and reuse custom circuit
def my_custom_gate(n_qubits):
    qc = QuantumCircuit(n_qubits, name='Custom')
    qc.h(range(n_qubits))
    for i in range(n_qubits-1):
        qc.cx(i, i+1)
    return qc.to_gate()

# Use in circuit
qc_main = QuantumCircuit(4)
qc_main.append(my_custom_gate(3), range(3))
print('Using custom gate:')
qc_main.draw('mpl')
```
